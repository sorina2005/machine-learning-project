```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_recall_fscore_support

SEED = 42
TARGET_COLUMN = 'Category'
TRAIN_TEST_RATIO = 0.2

np.random.seed(SEED)
```

```python
df = pd.read_csv('HepatitisCdata.csv')
df = df.drop(columns=['Unnamed: 0'])

X = df.drop(columns=[TARGET_COLUMN])
y = df[TARGET_COLUMN]

print(f"Dimensiunile datasetului: {df.shape[0]} randuri si {df.shape[1]} coloane")
print(f"Tipurile de variabile:\n{df.dtypes}")
```

```python
def load_data(path):
    df = pd.read_csv(path)
    if 'Unnamed: 0' in df.columns:
        df = df.drop(columns=['Unnamed: 0'])
    X = df.drop(columns=['Category'])
    y = df['Category']
    X['Sex'] = X['Sex'].map({'m': 1, 'f': 0})
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)
    return X, y_encoded, le

def preprocess_split(X, y, test_size, seed):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=seed, stratify=y
    )
    imputer = SimpleImputer(strategy='median')
    scaler = RobustScaler()
    
    X_train_proc = imputer.fit_transform(X_train)
    X_train_proc = scaler.fit_transform(X_train_proc)
    
    X_test_proc = imputer.transform(X_test)
    X_test_proc = scaler.transform(X_test_proc)
    
    return X_train_proc, X_test_proc, y_train, y_test
```

```python
def plot_results(y_true, y_pred, y_prob, labels):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=labels, yticklabels=labels)
    plt.title('Matrice de Confuzie')
    plt.show()

    plt.figure(figsize=(8, 6))
    for i in range(len(labels)):
        fpr, tpr, _ = roc_curve(y_true == i, y_prob[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'{labels[i]} (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.title('Curba ROC (One-vs-Rest)')
    plt.legend()
    plt.show()
```

```python
def run_experiment(model, X, y, test_size, seed, labels):
    X_train, X_test, y_train, y_test = preprocess_split(X, y, test_size, seed)
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)
    
    print(classification_report(y_test, y_pred, target_names=labels))
    plot_results(y_test, y_pred, y_prob, labels)
    
    return model
```
